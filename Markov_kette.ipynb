## Eigenschaften

- **Endliche Zustandsräume**: Meist wird mit einer endlichen Menge von Zuständen gearbeitet
- **Homogenität**: Die Übergangswahrscheinlichkeiten sind zeitunabhängig
- **Normalität**: Die Summe der Übergangswahrscheinlichkeiten in jeder Zeile ist 1

Mathematisch bedeutet dies, dass für jede Zeile $i$ der Übergangsmatrix gilt:

$$\sum_{j=1}^{n} p_{ij} = 1 \quad \text{für alle } i$$

Dies drückt aus, dass von jedem Zustand $i$ aus mit **Sicherheit** in einen der möglichen Zustände übergegangen wird. Die Wahrscheinlichkeiten aller möglichen nächsten Zustände summieren sich daher zu 1.
